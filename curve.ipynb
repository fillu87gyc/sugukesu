```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
A = np.arange(-np.pi, np.pi, 0.1)
B = np.sin(A)

plt.plot(A, B)
```

```python
a = 23
b = 32
r = 1
x_list =list()
y_list= list()
for i in range(1000):
    t = i / 1000 * 2 * np.pi
    x_list.append((r * np.sin(a * t)))
    y_list.append(-(r * np.cos(b * t)))
plt.plot(x_list,y_list)
```
